```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams

save_figures_to_latex = False
figures_path = '/home/tellocam/Projects/CSMCA/Exercise_8/latex_ex_8/figures'

rcParams['figure.figsize'] = 10, 7

plt.rc('text', usetex=True)
plt.rcParams['text.usetex']=True

plt.rc('font', family='sans-serif', weight='bold')

SMALL_SIZE = 20
MEDIUM_SIZE = 20
BIGGER_SIZE = 25

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rcParams['figure.dpi'] = 500
```

```python
data_2_rtx_ocl = np.loadtxt('ex8_2_rtx_OpenCL.txt', delimiter=',')
data_2_rtx_cuda = np.loadtxt('ex8_2_rtx_Cuda.txt', delimiter=',')
data_2_k40_ocl = np.loadtxt('ex8_2_k40_OpenCL.txt', delimiter=',')
data_2_k40_cuda = np.loadtxt('ex8_2_k40_Cuda.txt', delimiter=',')
data_3 = np.loadtxt('ex8_3.txt', delimiter=',')
```

```python
plt.figure(1)
plt.loglog(data_2_rtx_ocl[:,0],data_2_rtx_ocl[:,1], color='red', linewidth=2, label='OpenCL DOTP on RTX')
plt.loglog(data_2_rtx_cuda[:,0],data_2_rtx_cuda[:,1], color='red',marker="v", linewidth=2, label='CUDA DOTP on RTX')
plt.loglog(data_2_k40_ocl[:,0],data_2_k40_ocl[:,1], color='blue', linewidth=2, label='OpenCL DOTP on K40')
plt.loglog(data_2_k40_cuda[:,0],data_2_k40_cuda[:,1], color='blue',marker="v", linewidth=2, label='CUDA DOTP on K40')
plt.grid(True, which="both")
plt.xlabel('Vector Size $\mathrm{N}$ / -')
plt.ylabel('Elapsed time / s')
plt.legend()
plt.tight_layout()
if(save_figures_to_latex==True):
    plt.savefig(figures_path + '/task_8_2_plot.pdf')

plt.figure(2)
plt.loglog(data_2_k40_cuda[:,0],data_2_k40_cuda[:,1], color='red', linewidth=2, label='CUDA DOTP on K40')
plt.loglog(data_2_k40_ocl[:,0],data_2_k40_ocl[:,1], color='blue', linewidth=2, label='OpenCL DOTP on K40')
plt.loglog(data_3[:,0],data_3[:,1], color='green', linewidth=2, label='OpenCL DOTP on CPU')
plt.grid(True, which="both")
plt.xlabel('Vector Size $\mathrm{N}$ / -')
plt.ylabel('Elapsed time / s')
plt.legend()
plt.tight_layout()
if(save_figures_to_latex==True):
    plt.savefig(figures_path + '/task_8_3_plot.pdf')
```
