```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams

save_figures_to_latex = False
figures_path = '/home/tellocam/CSE/CSMCA/Exercise_9/latex_ex_9/figures'

rcParams['figure.figsize'] = 10, 7

plt.rc('text', usetex=True)
plt.rcParams['text.usetex']=True

plt.rc('font', family='sans-serif', weight='bold')

SMALL_SIZE = 20
MEDIUM_SIZE = 20
BIGGER_SIZE = 25

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rcParams['figure.dpi'] = 500
```

```python
k40_cuda = np.loadtxt('k40_cuda.txt', delimiter=',')
k40_hip = np.loadtxt('k40_hip.txt', delimiter=',')
rtx_cuda = np.loadtxt('rtx_cuda.txt', delimiter=',')
rtx_hip = np.loadtxt('rtx_hip.txt', delimiter=',')
```

```python
plt.figure(1)
plt.loglog(k40_cuda[:,0],k40_cuda[:,1], color='red', linewidth=2, label='CUDA CG on K40')
plt.loglog(k40_hip[:,0],k40_hip[:,1], color='red',marker="v", linewidth=2, label='HIP CG on K40')
plt.loglog(rtx_cuda[:,0],rtx_cuda[:,1], color='blue', linewidth=2, label='CUDA CG on RTX')
plt.loglog(rtx_hip[:,0],rtx_hip[:,1], color='blue',marker="v", linewidth=2, label='HIP CG on RTX')
plt.grid(True, which="both")
plt.xlabel('Points Per Direction $\sqrt{\mathrm{N}}$ / -')
plt.ylabel('Elapsed time / s')
plt.legend()
plt.tight_layout()
if(save_figures_to_latex==True):
    plt.savefig(figures_path + '/task_9_2_plot.pdf')
```
