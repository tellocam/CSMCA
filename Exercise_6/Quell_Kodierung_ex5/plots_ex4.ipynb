```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
rcParams['figure.figsize'] = 10, 7

plt.rc('text', usetex=True)
plt.rcParams['text.usetex']=True

plt.rc('font', family='sans-serif', weight='bold')

SMALL_SIZE = 20
MEDIUM_SIZE = 20
BIGGER_SIZE = 25

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rcParams['figure.dpi'] = 500  
```

```python
coding_on_camilos_laptop = False
figures_path = '/home/tellocam/CSE/CSMCA/Exercise_5/latex_ex_5/figures'

k40_5_1c = np.loadtxt('5_1_c_k40.txt', delimiter=',')
rtx_5_1c = np.loadtxt('5_1_c_rtx.txt', delimiter=',')

rtx_5_1d = np.loadtxt('5_1_d_rtx.txt', delimiter=',')

k40_5_1e = np.loadtxt('5_1_e_k40.txt', delimiter=',')
rtx_5_1e = np.loadtxt('5_1_e_rtx.txt', delimiter=',')

rtx_5_2c = np.loadtxt('5_2c.txt', delimiter=',')
```

```python
plt.figure(0)
plt.title('Task 1c - Practical Peak Memory Bandwidth')
plt.axhline(192.957, label='K40 max. = 192.957 GB/s', linestyle='dashed', color='red')
plt.axhline(337.089, label='RTX max. = 337.089 GB/s', linestyle='dashed', color='blue')
plt.loglog(k40_5_1c[:,0], k40_5_1c[:,1], label='K40 TESLA', color='red')
plt.loglog(rtx_5_1c[:,0], rtx_5_1c[:,1], label='RTX 3060', color='blue')
plt.xlabel('N / -')
plt.ylabel('Latency / $\mathrm{GB} \, \mathrm{s}^{-1}$')
plt.grid()
plt.legend()
if(coding_on_camilos_laptop==True):
    plt.savefig(figures_path + '/task_1_c.pdf')

plt.figure(1)
plt.title('Task 1d - Maximum number of atomicAdd()')
plt.axhline(6.07479e8, label='RTX max. = 337.089 atomicadd/s ', linestyle='dashed', color='blue')
plt.loglog(rtx_5_1d[:,0], rtx_5_1d[:,1], label='RTX 3060', color='blue')
plt.xlabel('N / -')
plt.ylabel('atomicAdds() per second /  $\mathrm{s}^{-1}$')
plt.grid()
plt.legend()
if(coding_on_camilos_laptop==True):
    plt.savefig(figures_path + '/task_1_d.pdf')

plt.figure(2)
plt.title('Task 1e - Peak Floating Point Rate for Doubles')
plt.axhline(2851.91, label='K40 max. = 2851.91 GFLOPs/s', linestyle='dashed', color='red')
plt.axhline(13419.8, label='RTX max. = 13419.8 GFLOPs/s', linestyle='dashed', color='blue')
plt.loglog(k40_5_1e[:,0], k40_5_1e[:,1], label='K40 TESLA', color='red')
plt.loglog(rtx_5_1e[:,0], rtx_5_1e[:,1], label='RTX 3060', color='blue')
plt.xlabel('N / -')
plt.ylabel('Floating Point Rate / GFLOPs $\mathrm{s}^{-1}$')
plt.grid()
plt.legend()
if(coding_on_camilos_laptop==True):
    plt.savefig(figures_path + '/task_1_e.pdf')

plt.figure(3)
plt.title('Task 2c - CG Time Convergence')
plt.loglog(rtx_5_2c[:,0], rtx_5_2c[:,1], label='K40 TESLA', color='red')
plt.xlabel('$\sqrt{\mathrm{N}}$ / -')
plt.ylabel('CG elapsed time / s')
plt.grid()
if(coding_on_camilos_laptop==True):
    plt.savefig(figures_path + '/task_2_c_1.pdf')

plt.figure(4)
plt.title('Task 2c - CG Iterations Convergence')
plt.loglog(rtx_5_2c[:,0], rtx_5_2c[:,2], label='RTX 3060', color='blue')
plt.xlabel('$\sqrt{\mathrm{N}}$ / -')
plt.ylabel('CG iterations / -')
plt.grid()
#plt.legend()
if(coding_on_camilos_laptop==True):
    plt.savefig(figures_path + '/task_2_c_2.pdf')
```
