```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
rcParams['figure.figsize'] = 10, 7

plt.rc('text', usetex=True)
plt.rcParams['text.usetex']=True

plt.rc('font', family='sans-serif', weight='bold')

SMALL_SIZE = 20
MEDIUM_SIZE = 20
BIGGER_SIZE = 25

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rcParams['figure.dpi'] = 500  
```

```python
coding_on_camilos_laptop = True
figures_path = '/home/tellocam/CSE/CSMCA/Exercise _3/latex_ex_3/figures'

data_1_a_1 = np.loadtxt('ex3_1a_1.txt', delimiter=',')
data_1_a_2 = np.loadtxt('ex3_1a_2.txt', delimiter=',')
data_1_b_1 = np.loadtxt('ex3_1b_1.txt', delimiter=',')
data_1_b_2 = np.loadtxt('ex3_1b_2.txt', delimiter=',')

plt.figure(1)
plt.title('Strided Memory Access - Task a)')
plt.semilogy(data_1_a_1[:,0],data_1_a_1[:,1], color='red', linewidth=2, label='Effective Bandwidth on K40')
plt.semilogy(data_1_a_2[:,0],data_1_a_2[:,1], color='blue', linewidth=2, label='Effective Bandwidth on RTX3060')
plt.grid(True, which="both")
plt.xlabel('$\mathrm{k}$ / -')
plt.ylabel('$\mathrm{BW}_{\mathrm{eff}}$ / $\mathrm{GB} \, \mathrm{s}^{-1}$')
plt.legend()
plt.tight_layout()
if(coding_on_camilos_laptop==True):
    plt.savefig(figures_path + '/task_1_a.pdf')

plt.figure(2)
plt.title('Offset Memory Access - Task b)')
plt.plot(data_1_b_1[:,0],data_1_b_1[:,1], color='red', linewidth=2, label='Effective Bandwidth on K40')
plt.plot(data_1_b_2[:,0],data_1_b_2[:,1], color='blue', linewidth=2, label='Effective Bandwidth on RTX3060')
plt.axvline(x=4, color = 'black', linestyle='dashed', label='show if $\mathrm{k} \, \mathrm{mod}(4) = 0$ ')
for i in range (5,64):
    if(i%4 == 0):
        plt.axvline(x=i, color = 'black', linestyle='dashed')
plt.grid(True, which="both")
plt.xlabel('$\mathrm{k}$ / -')
plt.ylabel('$\mathrm{BW}_{\mathrm{eff}}$ / $\mathrm{GB} \, \mathrm{s}^{-1}$')
plt.legend()
plt.tight_layout()
if(coding_on_camilos_laptop==True):
    plt.savefig(figures_path + '/task_1_b.pdf')

```
