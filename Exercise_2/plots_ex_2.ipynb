```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

plt.rc('text', usetex=True)
plt.rcParams['text.usetex']=True

plt.rc('font', family='sans-serif', weight='bold')

SMALL_SIZE = 20
MEDIUM_SIZE = 20
BIGGER_SIZE = 25

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rcParams['figure.dpi'] = 500  
```

```python
coding_on_camilos_laptop = True

data_1_a = np.loadtxt('ex_2_1_a.txt', delimiter=',')
data_1_b1 = np.loadtxt('ex_2_1_b1.txt', delimiter=',')
data_1_b2 = np.loadtxt('ex_2_1_b2.txt', delimiter=',')
data_1_d = np.loadtxt('ex_2_1_d.txt', delimiter=',')
data_1_e = np.loadtxt('ex_2_1_e.txt', delimiter=',')
data_2_a = np.loadtxt('ex_2_2_a.txt', delimiter=',')
data_2_b = np.loadtxt('ex_2_2_b.txt', delimiter=',')

plt.figure(1)
plt.title('Basic Cuda Task a')
plt.loglog(data_1_a[:,0],data_1_a[:,1], color='red', linewidth=2, label='Time for cudaMalloc()')
plt.loglog(data_1_a[:,0],data_1_a[:,2], color='blue', linewidth=2, label='Time for cudaFree()')
plt.grid(True, which="both")
plt.xlabel('$\mathrm{N}$ / -')
plt.ylabel('Time / $\mathrm{s}$')
plt.legend()
plt.tight_layout()
if(coding_on_camilos_laptop==True):
    plt.savefig('/home/tellocam/CSE/CSMCA/Exercise_2/latex_ex_2/figures/task_1_a.pdf')

plt.figure(2)
plt.title('Basic Cuda Task b')
plt.loglog(data_1_b1[:,0],data_1_b1[:,1], color='red', linewidth=2, label='Initialized in Cuda Kernel')
plt.loglog(data_1_b2[:,0],data_1_b2[:,1], color='blue', linewidth=2, label='Initialized with cudaMemcpy()')
plt.grid(True, which="both")
plt.xlabel('$\mathrm{N}$ / -')
plt.ylabel('Time / $\mathrm{s}$')
plt.legend()
plt.tight_layout()
if(coding_on_camilos_laptop==True):
    plt.savefig('/home/tellocam/CSE/CSMCA/Exercise_2/latex_ex_2/figures/task_1_b.pdf')

plt.figure(3)
plt.title('Basic Cuda Task d')
plt.loglog(data_1_d[:,0],data_1_d[:,1], color='red', linewidth=2, label='gpu_x + gpu_y = gpu_z')
plt.grid(True, which="both")
plt.xlabel('$\mathrm{N}$ / -')
plt.ylabel('Time / $\mathrm{s}$')
plt.legend()
plt.tight_layout()
if(coding_on_camilos_laptop==True):
    plt.savefig('/home/tellocam/CSE/CSMCA/Exercise_2/latex_ex_2/figures/task_1_d.pdf')


plt.figure(4)
plt.title('Basic Cuda Task e')
plt.loglog(data_1_e[:,0],data_1_e[:,1], color='blue', linewidth=2, label='Square Gridsizes')
plt.grid(True, which="both")
plt.xlabel('$\sqrt{\mathrm{Trds \cdot Blks}}$ / -')
plt.ylabel('Time / $\mathrm{s}$')
plt.legend()
plt.tight_layout()
if(coding_on_camilos_laptop==True):
    plt.savefig('/home/tellocam/CSE/CSMCA/Exercise_2/latex_ex_2/figures/task_1_e.pdf')

plt.figure(5)
plt.title('Dot Product Task a and b')
plt.loglog(data_2_a[:,0],data_2_a[:,1], color='red', linewidth=2, label='2 GPU Stages')
plt.loglog(data_2_b[:,0],data_2_b[:,1], color='blue', linewidth=2, label='CPU and GPU Stage')
plt.grid(True, which="both")
plt.xlabel('$\mathrm{N}$ / -')
plt.ylabel('Time / $\mathrm{s}$')
plt.legend()
plt.tight_layout()
if(coding_on_camilos_laptop==True):
    plt.savefig('/home/tellocam/CSE/CSMCA/Exercise_2/latex_ex_2/figures/dot_product.pdf')
```
